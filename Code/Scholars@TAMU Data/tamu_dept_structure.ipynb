## Prep Work

```python
with open('../../Data/Scholars@TAMU/my_api_calls/general_data.json', 'r') as infile:
    apis = json.load(infile)
```

```python
position_orgs = []

count = {'no_positions': 0, 'no_orgs': 0}
for api in apis:
    try:
        for position in api['positions']:
            try:
                for org in position['organizations']:
                    position_orgs.append({org['label']: [parent['label'] for parent in org['parent']]})
            except:
                count['no_orgs'] += 1
    except:
        count['no_orgs'] += 1

count
```
